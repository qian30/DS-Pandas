```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
nog = [0, 0, 0, 0, 0, 0]
for i in range(30):
    if(df['High school system'][i] == 'National' and df['Gender'][i] == 'male'):
        nog [0] += df['Num of graduates'][i]
    elif (df['High school system'][i] == 'National' and df['Gender'][i] == 'female'):
        nog [1] += df['Num of graduates'][i]
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'male'):
        nog [2] += df['Num of graduates'][i]
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'female'):
        nog [3] += df['Num of graduates'][i]
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'male'):
        nog [4] += df['Num of graduates'][i]
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'female'):
        nog [5] += df['Num of graduates'][i]
print(nog)
```

```python
pu = [0, 0, 0, 0, 0, 0]
for i in range(30):
    if(df['High school system'][i] == 'National' and df['Gender'][i] == 'male'):
        pu [0] += df['Num of enroll in police university'][i]
    elif (df['High school system'][i] == 'National' and df['Gender'][i] == 'female'):
        pu [1] += df['Num of enroll in police university'][i]
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'male'):
        pu [2] += df['Num of enroll in police university'][i]
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'female'):
        pu [3] += df['Num of enroll in police university'][i]
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'male'):
        pu [4] += df['Num of enroll in police university'][i]
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'female'):
        pu [5] += df['Num of enroll in police university'][i]
print(pu)
```

```python
pc = [0, 0, 0, 0, 0, 0]
for i in range(30):
    if(df['High school system'][i] == 'National' and df['Gender'][i] == 'male'):
        pc [0] += df['Num of enroll in police college'][i]
    elif (df['High school system'][i] == 'National' and df['Gender'][i] == 'female'):
        pc [1] += df['Num of enroll in police college'][i]
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'male'):
        pc [2] += df['Num of enroll in police college'][i]
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'female'):
        pc [3] += df['Num of enroll in police college'][i]
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'male'):
        pc [4] += df['Num of enroll in police college'][i]
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'female'):
        pc [5] += df['Num of enroll in police college'][i]
print(pc)
```

```python
ms = [0, 0, 0, 0, 0, 0]
count = [0, 0, 0, 0, 0, 0]
ams = [0, 0, 0, 0, 0, 0]
for i in range(30):
    if(df['High school system'][i] == 'National' and df['Gender'][i] == 'male'):
        ms [0] += df['Num of enroll in military school'][i]
        count[0] += 1
    elif (df['High school system'][i] == 'National' and df['Gender'][i] == 'female'):
        ms [1] += df['Num of enroll in military school'][i]
        count[1] += 1
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'male'):
        ms [2] += df['Num of enroll in military school'][i]
        count[2] += 1
    elif (df['High school system'][i] == 'county & city' and df['Gender'][i] == 'female'):
        ms [3] += df['Num of enroll in military school'][i]
        count[3] += 1
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'male'):
        ms [4] += df['Num of enroll in military school'][i]
        count[4] += 1
    elif (df['High school system'][i] == 'private' and df['Gender'][i] == 'female'):
        ms [5] += df['Num of enroll in military school'][i]
        count[5] += 1
for i in range (6):
    ams[i] = round(ms[i]/count[i], 2)
print(ams)
```

```python
puer = [0, 0, 0, 0, 0, 0]
for i in range(6):
    puer[i] = round((pu[i]/nog[i])*100, 2)
    puer[i] = str(puer[i])+'%'
print(puer)
```

```python
pcer = [0, 0, 0, 0, 0, 0]
for i in range(6):
    pcer[i] = round((pc[i]/nog[i])*100, 2)
    pcer[i] = str(pcer[i])+'%'
print(pcer)
```

```python
mser = [0, 0, 0, 0, 0, 0]
for i in range(6):
    mser[i] = round((ms[i]/nog[i])*100, 2)
    mser[i] = str(mser[i])+'%'
print(mser)
```

```python
ans = pd.DataFrame({
    'High School System' : ['National', 'National', 'country & city', 'country & city', 'private', 'private'],
    'Gender' : ['male', 'female', 'male', 'female','male', 'female'],
    'Num of Graduates' : nog,
    'Enroll in Police University' : pu,
    'Enroll in Police College' : pc,
    'Average Enroll in Military' : ams,
    'Police Univ Enrollment Rate' : puer,
    'Police College Enrollment Rate' : pcer,
    'Military School Enrollment Rate' : mser
})
ans
```
